```python
import numpy as np
import h5py
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.interpolate import interp1d
import torch
from sklearn.model_selection import train_test_split
```

## Prepare the Dataset

```python
filename = "../data/"
raw_data = h5py.File(filename+'ViSAPy_nonfiltered.h5','r')
raw_data = raw_data['data'][()]
print(raw_data.shape)
```

```python
train_data, test_data = train_test_split(raw_data, test_size=0.22, random_state=5132, shuffle=True)
print(train_data.shape, test_data.shape)
```

## Saving/Loading Train and Test Data

```python
np.save('train_data_101', train_data)
np.save('test_data_101', test_data)

#train_data = np.load('train_data_101.npy')
#test_data = np.load('test_data_101.npy')
```
